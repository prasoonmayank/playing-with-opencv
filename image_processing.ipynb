```python
import numpy as np
import imutils
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("images/inspirational-motivational-quotes-24.jpg")
plt.imshow(image)
```

```python
M = np.float32([[1,0,25],[0,1,50]])
shifted = cv2.warpAffine(image,M,(image.shape[1],image.shape[0]))
plt.imshow(shifted)
```

```python
M = np.float32([[1,0,-50],[0,1,-90]])
shifted = cv2.warpAffine(image,M,(image.shape[1],image.shape[0]))
plt.imshow(shifted)
```
