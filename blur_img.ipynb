```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("images/nelsonmandela1-2x.jpg")
plt.imshow(image)
```

```python
blurred = np.hstack([
    cv2.blur(image,(3,3)),
    cv2.blur(image,(5,5)),
    cv2.blur(image,(7,7))
])
plt.imshow(blurred)
```

```python
blurred33 = np.hstack([cv2.blur(image,(3,3))])
plt.imshow(blurred33)
```

```python
blurred1 = np.hstack([cv2.blur(image,(11,11))])
plt.imshow(blurred1)
```

```python
gaussianblur = np.hstack([cv2.GaussianBlur(image,(11,11),0)])
plt.imshow(gaussianblur)
```

```python
gaussianblur = np.hstack([cv2.GaussianBlur(image,(3,3),0)])
plt.imshow(gaussianblur)
```

```python
gaussianblur = np.hstack([cv2.GaussianBlur(image,(19,19),0)])
plt.imshow(gaussianblur)
```
