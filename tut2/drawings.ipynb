```python
import numpy as np
import cv2

%matplotlib inline
import matplotlib.pyplot as plt

canvas = np.zeros((300,300,3), dtype="uint8")
```

```python
green = (0,255,0)
cv2.line(canvas,(0,0),(300,300),green)
plt.imshow(canvas)
```

```python
blue = (0,0,255)
red = (255,0,0)
cv2.line(canvas,(300,0),(0,300),red,1)
cv2.line(canvas,(150,0),(150,300),blue)
plt.imshow(canvas)
```

```python
cv2.rectangle(canvas,(10,10),(60,60),red)
plt.imshow(canvas)
```
